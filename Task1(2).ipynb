```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.exponential(size=(4,4), scale=1.0)
```

```python
# Create a random 100000x1 array from exponential distribution
exponential = np.random.exponential(size=(100000), scale=1.0)
```

```python
# Create a random 100000x1 array from normal distribution
normal =np.random.normal(size=(100000))
```

```python
# Create a random 100000x1 array from uniform distribution
uniform =np.random.rand(100000)
```

```python
#exponential distribution
plt.hist(exponential,density=True,bins=50,histtype="step",color="blue",label="Exponential")
#normal distribution
plt.hist(normal,density=True,bins=50,histtype="step",color="red",label="Normal")
#uniform distribution
plt.hist(uniform,density=True,bins=50,histtype="step",color="green",label="Uniform")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
# Create a grid of X and Y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
```

```python
# Calculate Z values
Z = X**2 + Y**2
```

```python
# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Z = X^2 + Y^2')

# Show the plot
plt.show()
```

```python
#4
import pandas as pd

!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv

df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
df.head()
```

```python
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
data = df[features]
```

```python
pearson_corr = data.corr(method='pearson')
spearman_corr = data.corr(method='spearman')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 8))

# Plot Pearson Correlation Heatmap
sns.heatmap(pearson_corr, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1, ax=axes[0])
axes[0].set_title('Pearson Correlation')
axes[0].set_xticklabels(pearson_corr.columns, rotation=45)
axes[0].set_yticklabels(pearson_corr.columns, rotation=0)

# Plot Spearman Correlation Heatmap
sns.heatmap(spearman_corr, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1, ax=axes[1])
axes[1].set_title('Spearman Rank Correlation')
axes[1].set_xticklabels(spearman_corr.columns, rotation=45)
axes[1].set_yticklabels(spearman_corr.columns, rotation=0)

plt.tight_layout()
plt.show()
```
